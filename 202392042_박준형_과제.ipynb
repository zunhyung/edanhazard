```python
#과제
from PIL import Image
import matplotlib.pyplot as plt

im = Image.open('./images/parrot.png')
im2 = Image.open('./images/apple.png')
im3 = Image.open('./images/birds.png')
im4 = Image.open('./images/butterfly.png')

plt.subplot(221)
plt.imshow(im)
plt.yticks([0, 50, 100, 150, 200, 250, 300])

plt.subplot(222)
plt.imshow(im2)
plt.yticks([0, 50, 100, 150, 200, 250, 300])

plt.subplot(223)
plt.imshow(im3)
plt.xticks([0, 100, 200, 300, 400, 500])

plt.subplot(224)
plt.imshow(im4)
plt.xticks([0, 100, 200, 300, 400, 500])

plt.show()
```
